```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query1 = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result
hygiene_score_eq_20_documents = establishments.count_documents(query1)
print("Number of documents in which the hygiene score is equal to 20:", hygiene_score_eq_20_documents)
print("Here's the first document:","\n")

# Display the first document in the results using pprint
results1a = establishments.find(query1)     # for use below
results1b = establishments.find_one(query1)
pprint(results1b)
```

```python
# Convert the result to a Pandas DataFrame
import pandas as pd
from pandas import json_normalize
json_results_df1 = pd.json_normalize(results1a)

# Display the number of rows in the DataFrame
print("Number of Rows in the DataFrame:", len(json_results_df1))

# Display the first 10 rows of the DataFrame
print("Here's the first 10 rows:","\n")
json_results_df1.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
# Note: A review in MongoDB Compass reveals no "London" but 37 entires of "City of London Corporation" which will
# be considered as "London"
query2 = {'LocalAuthorityName': {"$regex" : "London"},
         'RatingValue': {'$gte': 4}
         }

# Use count_documents to display the number of documents in the result
london_rating_value_gte_4 = establishments.count_documents(query2)
print("Number of documents in London with a rating value greater than or equal to 4:", london_rating_value_gte_4)
print("Here's the first document:","\n")

# Display the first document in the results using pprint
results2a = establishments.find(query2)     # for use below
results2b = establishments.find_one(query2)
pprint(results2b)
```

```python
# Convert the result to a Pandas DataFrame
json_results_df2 = pd.json_normalize(results2a)

# Display the number of rows in the DataFrame
print("Number of Rows in the DataFrame:", len(json_results_df2))

# Display the first 10 rows of the DataFrame
print("Here's the first 10 rows:","\n")
json_results_df2.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Encountered difficulties with Decimal128; after many hours of effort unable to find a solution in which it is not necessary
# to determine the values of latitude and longitude. It is necessary to perform the followiong to find
# the values of latitude and longitude of 'Penang Flavours' so that they may be plugged into the variables provided.

latitude = establishments.find_one({"BusinessName" : "Penang Flavours"}, {"geocode.latitude"})
longitude = establishments.find_one({"BusinessName" : "Penang Flavours"}, {"geocode.longitude"})
pprint(latitude)
pprint(longitude)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

# changing to "query" to "query3" be cpmsistemt with above
# "within 0.01 degree" is interpreted as not including 0.01, i.e., $lte and $gte are not used 
query3 = {"$and":[
    {'RatingValue': 5}, 
    {'geocode.latitude': {'$lte': (latitude + degree_search), '$gte': (latitude - degree_search)}}, 
    {'geocode.longitude': {'$lte': (longitude + degree_search), '$gte': (longitude - degree_search)}} 
    ]}
sort = [('scores.Hygiene', 1)] 

# Limiting the output of the pprint to BusinessName and geocode
fields = ['BusinessName', 'geocode']
results3 = list(establishments.find(query3, fields).sort(sort).limit(5))

# Print the results
print("Here are the top 5 establishments meeting the stated criteria:","\n")
pprint(results3)
```

```python
# Convert result to Pandas DataFrame
print("Here is the Pandas DataFrame of top 5 establishments meeting the stated criteria:","\n")
json_results_df3 = pd.json_normalize(results3)
json_results_df3.head(5)
```
